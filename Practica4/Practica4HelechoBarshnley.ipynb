```python
import numpy 
from numpy import *
from random import randint #Se importan las librerias de numpy
```

```python
def F(var,coef): #Defino la función F que realizará la operación entre vector y matriz
    M = (coef[:4]).reshape(2,2) # Construyo la matriz de operacioón con base en un vector (lista en array)
    Vector = coef[4:6] #Construyo el vector a sumar de la misma forma usando los indices de la lista
    xop = M[0,0]*var[0] + M[0,1]*var[1] #Se realiza la operación vector-matriz de forma manual
    yop = M[1,0]*var[0] + M[1,1]*var[1]
    vecop = array([xop,yop]) #obtengo el primer vector resultante el cual será sumado al independiente en el resultado r
    r = vecop + Vector
    return r
```

```python
def chance(): #Defino la función chance como l función a generar los coeficientes aleatoreamente
    import random #Se importa la libreria random y defino a n como un número entero al azar entre 1 y 100
    n = random.randint(1,100)
    a,b,c,d,e,f = 1 ,2,3,4,5,6 #Les asigno valores arbitrarios a las variables de los coeficientes para poder definirlas
    coef = [] #Creo una lista coef vacía
    if n == 1: #Se crea el bucle if dependiendo del valor asignado al azar de n en donde se le reasignarán valores a las variables
        a = float(0) #Finalmente cada condicional añade las variables a la lista coef
        b = float(0)
        c = float(0)
        d = float(0.16)
        e = float(0)
        f = float(0)
        coef.append(a)
        coef.append(b)
        coef.append(c)
        coef.append(d)
        coef.append(e)
        coef.append(f)
    if 2 <= n <= 86:
        a = float(0.85)
        b = float(0.04)
        c = float(-0.04)
        d = float(0.85)
        e = float(0)
        f = float(1.6)
        coef.append(a)
        coef.append(b)
        coef.append(c)
        coef.append(d)
        coef.append(e)
        coef.append(f)
    if 87 <= n <= 93:
        a = float(0.2)
        b = float(-0.26)
        c = float(0.23)
        d = float(0.22)
        e = float(0)
        f = float(1.6)
        coef.append(a)
        coef.append(b)
        coef.append(c)
        coef.append(d)
        coef.append(e)
        coef.append(f)
    if 94 <= n <= 100:
        a = float(-0.15)
        b = float(0.28)
        c = float(0.26)
        d = float(0.24)
        e = float(0)
        f = float(0.44)
        coef.append(a)
        coef.append(b)
        coef.append(c)
        coef.append(d)
        coef.append(e)
        coef.append(f)
    coefnew = array(coef) #Transformo la lista coef en un vector coefnew el cual será la salida de la función
    return coefnew
        
```

```python
chance() #Prueba de la función chance
```

```python
def Barshnley(n): #Defino a la función Barshnley la cual tomará como entrada el número de iteraciones
    xy = [0,0] #Establezco la semilla con el punto 0,0
    coef0 = chance() #Obtengo una combinación al azar de vectores
    resnew = F(xy, coef0) #Defino al resultado inicial de aplicar la funcion F
    i = 1 #Establezco un contador
    xp = [] #Defino  a la lista de componentes x de todas las iteraciones
    yp = [] #Defino a la lista de componentes y de todas las iteraciones
    while i <= n: #Se inicia un bucle while que circule hasta el número de iteraciones deseadas
        coef = chance() #Obtengo una nueva combinación aleatoria de coeficientes
        resnew = F(resnew, coef) #Hago reiterativo la obtención del resultado tomando el anterior y aplicando la función F
        xs = resnew[0] #Defino una variable temporal con la primera entrada del resultado
        ys = resnew[1] #Defino una variable temporal con la segunda entrada del resultado
        xp.append(xs) #Se añaden las componentes x e y a sus respectivas listas
        yp.append(ys)
        
        i += 1 #El bucle continua
    import matplotlib.pyplot as plt #Se importan las librerias para graficar
    plt.scatter(xp, yp, s = 0.2, edgecolor ='green') #Defino los puntos a graficar como las listas de las componentes x e y de todas las iteraciones
    return
```

```python
Barshnley(15000) #Prueba de la función Barshnley a 15000 iteraciones
```

```python
def Rotation(n,var): #Defino la función de rotación de vectores que tomará un vector y el  número de grados a rotar
    a = (n * numpy.pi)/180 #Transformo los grados a radianes
    matr = array([cos(a),-sin(a),sin(a),cos(a)]) #Construyo un vector con los coeficientes de la matriz de rotación, evaluando los senos y cosenos en el ángulo radian deseado
    M = matr.reshape(2,2) #Construyo la matriz de rotación con el comando reshape obteniendo la matriz de 2x2
    xop = M[0,0]*var[0] + M[0,1]*var[1] #Se realiza la operación entre vector y matriz de forma manual
    yop = M[1,0]*var[0] + M[1,1]*var[1]
    vecop = array([xop,yop])
    return vecop #asignamos la salida como un vector, el vector original bajo la rotación
```

```python
def BarshnleyRot(n,a): #Finalmente definimos a la función BarshnleyRot como la generadora del helecho rotado
    xy = [0,0] #Mismo principio que en la función anterior
    coef0 = chance()
    resnew = F(xy, coef0)
    i = 1
    xp = []
    yp = []
    f = open("Barshnley.txt","w") #Se crea un archivo Barshnley.txt donde se escribirán los datos
    
    while i <= n:
        coef = chance()
        resnew = F(resnew, coef)
        resnew = Rotation(a,resnew) #Añadimos un paso adicional en el cual se rotará el vector resultante
        xs = resnew[0]
        ys = resnew[1]
        xp.append(xs)
        yp.append(ys)
        f.write("Componente x , Componente y \n") #Se da la instrucción de escribir sobre el archivo txt
        f.write("%f %f \n" %(xs,ys)) 
        
        i += 1 
    f.close #Se cierra el archivo
    
    import matplotlib.pyplot as plt
    plt.scatter(xp, yp, s = 0.2, edgecolor ='green')
    return
```
